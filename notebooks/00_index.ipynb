```python
import importlib.util, pathlib, sys

sdk_path = pathlib.Path('../sdk/multigen')
modules = sorted(p.stem for p in sdk_path.glob('*.py')
                 if not p.stem.startswith('_'))

print(f'SDK path: {sdk_path.resolve()}')
print(f'Total modules: {len(modules)}')
print('Modules:', ', '.join(modules))

# Spot-check a few imports
spot_check = ['tenancy', 'security', 'rag', 'knowledge', 'prompt_registry',
              'llm_adapter', 'simulator', 'lifecycle', 'feedback_loop']
print('\nSpot-check imports:')
for name in spot_check:
    path = sdk_path / f'{name}.py'
    spec = importlib.util.spec_from_file_location(name, path)
    mod = importlib.util.module_from_spec(spec)
    spec.loader.exec_module(mod)
    print(f'  OK: {name:25s} {len(mod.__all__)} exports')
```

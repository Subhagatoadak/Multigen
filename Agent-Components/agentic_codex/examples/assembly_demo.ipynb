# Assembly Coordinator Demo

```python
from agentic_codex import Agent, Context, Stage, AssemblyCoordinator, FunctionAdapter
from agentic_codex.core.schemas import AgentStep, Message

def writer_step(ctx: Context) -> AgentStep:
    return AgentStep(out_messages=[Message(role="assistant", content="draft")])

writer = Agent(writer_step, name="writer", role="writer")
coordinator = AssemblyCoordinator([Stage("write", writer)])
result = coordinator.run(goal="demo", inputs={})
result.messages[-1].content
```

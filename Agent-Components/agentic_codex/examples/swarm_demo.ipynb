# Swarm Coordinator Demo

```python
from agentic_codex import Agent, Context, SwarmCoordinator
from agentic_codex.core.schemas import AgentStep, Message

def explorer(name):
    def _step(ctx: Context) -> AgentStep:
        return AgentStep(out_messages=[Message(role="assistant", content=f"{name} idea")])
    return Agent(_step, name=name, role="explorer")

swarm = SwarmCoordinator([explorer("A"), explorer("B")])
result = swarm.run(goal="demo", inputs={})
[m.content for m in result.messages]
```
